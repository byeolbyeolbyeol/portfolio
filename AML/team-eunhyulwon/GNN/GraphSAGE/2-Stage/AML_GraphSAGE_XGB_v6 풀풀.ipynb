# 3) Timestamp 파싱 + 정렬

```python
# ── Timestamp -> ts (Datetime) ─────────────────────────────────────────────────
total_rows = q_master.select(pl.len()).collect().item()

q_master = q_master.with_columns(
    pl.col("Timestamp")
      .str.strptime(pl.Datetime, format="%Y/%m/%d %H:%M", strict=False)
      .alias("ts")
)

null_ts = q_master.filter(pl.col("ts").is_null()).select(pl.len()).collect().item()
print(f"Total rows    : {total_rows:,}")
print(f"NULL ts (skip): {null_ts:,}")
print(f"Valid ts      : {total_rows - null_ts:,}")

q_valid = q_master.filter(pl.col("ts").is_not_null())
```

# 4) Time-based Split (60 / 20 / 20, ref_ts 기준 · 누수 없음)

```python
# ── 경계 ts 계산 ──────────────────────────────────────────────────────────────
n = q_valid.select(pl.len()).collect().item()
q_sorted = q_valid.sort("ts")

train_cut_ts = q_sorted.select("ts").slice(int(n * 0.6), 1).collect().item()
val_cut_ts   = q_sorted.select("ts").slice(int(n * 0.8), 1).collect().item()
print(f"Train cutoff : {train_cut_ts}")
print(f"Val   cutoff : {val_cut_ts}")

# ── split 컬럼 부여 ───────────────────────────────────────────────────────────
q_split = q_sorted.with_columns(
    pl.when(pl.col("ts") <= train_cut_ts).then(pl.lit("train"))
      .when(pl.col("ts") <= val_cut_ts).then(pl.lit("val"))
      .otherwise(pl.lit("test"))
      .alias("split")
)

split_dist = (
    q_split.group_by(["split", "Is Laundering"])
           .agg(pl.len().alias("rows"))
           .collect()
)
print(split_dist)
```

# 5) 피처 엔지니어링

## 5-a) bucket_ts + 기본 파생 피처

```python
# ── 상수 / 룩업 ───────────────────────────────────────────────────────────────
HIGH_RISK_SENDER_BANKS = [
    "National Bank of Dallas", "Savings Bank of Augusta", "China Bank #27",
    "India Bank #96", "Brazil Bank #128", "National Bank of Milford",
    "Savings Bank of Sacramento", "Saudi Arabia Bank #14", "Israel Bank #16",
    "Golden Credit Union",
]
HIGH_RISK_RECEIVER_BANKS = [
    "China Bank #292", "China Bank #27", "China Bank #22", "Japan Bank #143",
    "Brazil Bank #50", "Bank of Denver", "Saudi Arabia Bank #56",
    "Israel Bank #48", "The Pine Bank", "National Bank of Milford",
]
ENTITY_TYPE_SCORE_MAP = {"Corporation": 2.0, "Sole Proprietorship": 1.5}
PAYMENT_RISK_MAP = {
    "ACH": 3.0, "Bitcoin": 2.5, "Cheque": 2.5,
    "Cash": 1.5, "Credit Card": 1.3, "Wire": 1.0, "Reinvestment": 1.0,
}

# ── bucket_ts (1h truncate) ───────────────────────────────────────────────────
q_feat = q_split.with_columns(
    pl.col("ts").dt.truncate("1h").alias("bucket_ts")
).with_columns(
    pl.col("bucket_ts").dt.strftime("%Y-%m-%d %H:%M:%S").alias("bucket_ts_str")
)

# ── Timestamp 파생 ────────────────────────────────────────────────────────────
q_feat = q_feat.with_columns([
    pl.col("ts").dt.hour().cast(pl.Int16).alias("hour"),
    pl.col("ts").dt.weekday().cast(pl.Int8).alias("day_of_week"),
    pl.col("ts").dt.weekday().is_in([6, 7]).alias("is_weekend"),
    pl.col("ts").dt.date().alias("ts_day"),
    pl.col("ts").dt.hour().is_between(0, 5).alias("is_dawn"),
    pl.when(pl.col("ts").dt.hour().is_between(0, 5)).then(pl.lit("dawn"))
      .when(pl.col("ts").dt.hour().is_between(6, 11)).then(pl.lit("morning"))
      .when(pl.col("ts").dt.hour().is_between(12, 17)).then(pl.lit("afternoon"))
      .otherwise(pl.lit("evening"))
      .alias("timeofday_bucket"),
]).with_columns([
    (pl.col("hour").cast(pl.Float32) * (2 * np.pi / 24)).sin().alias("hour_sin"),
    (pl.col("hour").cast(pl.Float32) * (2 * np.pi / 24)).cos().alias("hour_cos"),
    pl.col("day_of_week").cast(pl.Utf8).cast(pl.Categorical).alias("dow_cat"),
])

# ── Amount 파생 ───────────────────────────────────────────────────────────────
q_feat = q_feat.with_columns([
    pl.col("Amount_Paid_USD").cast(pl.Float32),
    pl.col("Amount_Received_USD").cast(pl.Float32),
    pl.col("Amount_Paid_USD").log1p().alias("log_amount_paid_usd"),
    pl.col("Amount_Received_USD").log1p().alias("log_amount_received_usd"),
    (pl.col("Amount_Paid_USD") % 1000  == 0).alias("is_round_1000_paid"),
    (pl.col("Amount_Received_USD") % 1000 == 0).alias("is_round_1000_received"),
    (pl.col("Amount_Paid_USD") % 10000 == 0).alias("is_round_10000_paid"),
])

# ── Payment Format 원핫 + 파생 ────────────────────────────────────────────────
q_feat = q_feat.with_columns([
    (pl.col("Payment Format") == "ACH").alias("is_ach"),
    (pl.col("Payment Format") == "Cheque").alias("is_cheque"),
    (pl.col("Payment Format") == "Bitcoin").alias("is_bitcoin_fmt"),
    (pl.col("Payment Format") == "Cash").alias("is_cash"),
    (pl.col("Payment Format") == "Credit Card").alias("is_credit_card"),
    (pl.col("Payment Format") == "Wire").alias("is_wire"),
    (pl.col("Payment Format") == "Reinvestment").alias("is_reinvestment"),
    ((pl.col("Payment Format") == "Bitcoin") | (pl.col("Payment Currency") == "Bitcoin"))
      .alias("is_crypto_transfer"),
    ((pl.col("Payment Format") == "ACH") & (pl.col("Amount_Paid_USD") >= 1_000_000))
      .alias("is_high_value_ach"),
    pl.col("Payment Format").replace(PAYMENT_RISK_MAP, default=1.0)
      .cast(pl.Float32).alias("payment_method_risk"),
])
q_feat = q_feat.with_columns([
    (pl.col("payment_method_risk") * pl.col("log_amount_paid_usd")).alias("risk_x_log_paid"),
    (pl.col("is_ach").cast(pl.Int8) * pl.col("log_amount_paid_usd")).alias("ach_x_log_paid"),
    (pl.col("is_bitcoin_fmt").cast(pl.Int8) * pl.col("log_amount_paid_usd")).alias("btc_x_log_paid"),
])

# ── Entity Type + 점수 ────────────────────────────────────────────────────────
q_feat = q_feat.with_columns([
    pl.col("Sender_Entity").cast(pl.Utf8).str.replace(r"\s*#\d+.*$", "").alias("sender_entity_type"),
    pl.col("Receiver_Entity").cast(pl.Utf8).str.replace(r"\s*#\d+.*$", "").alias("receiver_entity_type"),
]).with_columns([
    pl.col("sender_entity_type").replace(ENTITY_TYPE_SCORE_MAP, default=1.0)
      .cast(pl.Float32).alias("sender_entity_type_score"),
    pl.col("receiver_entity_type").replace(ENTITY_TYPE_SCORE_MAP, default=1.0)
      .cast(pl.Float32).alias("receiver_entity_type_score"),
])

# ── (F) bank 관련 ─────────────────────────────────────────────────────────────
q_feat = q_feat.with_columns([
    pl.col("Sender_Bank_Name").is_in(HIGH_RISK_SENDER_BANKS).alias("high_risk_sender_bank_flag"),
    pl.col("Receiver_Bank_Name").is_in(HIGH_RISK_RECEIVER_BANKS).alias("high_risk_receiver_bank_flag"),
])

q_feat = q_feat.with_columns([
    pl.col("From Bank").cast(pl.Utf8).cast(pl.Categorical),
    pl.col("To Bank").cast(pl.Utf8).cast(pl.Categorical),
])

q_feat = q_feat.with_columns([
    pl.col("Sender_Bank_Branch_ID").cast(pl.Utf8).cast(pl.Categorical),
    pl.col("Receiver_Bank_Branch_ID").cast(pl.Utf8).cast(pl.Categorical),
])

print("기본 피처 생성 완료. 현재 컬럼 수:", len(q_feat.columns))

# ── 5-a 끝: LazyFrame → DataFrame (eager) 전환 ──────────────────────────────
print("Collecting q_feat to eager DataFrame ...")
df_feat = q_feat.collect()
print(f"  shape: {df_feat.shape}")
```

## 5-b) Window Rolling 피처 (Amount 기반, shift(1) 누수방지)

```python
# ============================================================
# 5-b) Window Rolling 피처 (eager DataFrame 기반)
# ============================================================
df_feat = df_feat.with_columns(pl.lit(1).cast(pl.Int32).alias("_one"))

# sender 기준
df_feat = df_feat.sort(["From Account", "ts"])
df_feat = df_feat.with_columns([
    pl.col("Amount_Paid_USD")
      .rolling_sum_by("ts", window_size="1h").shift(1).over("From Account")
      .alias("s_wout_1h"),
    pl.col("Amount_Paid_USD")
      .rolling_sum_by("ts", window_size="24h").shift(1).over("From Account")
      .alias("s_wout_24h"),
    pl.col("Amount_Received_USD")
      .rolling_sum_by("ts", window_size="1h").shift(1).over("From Account")
      .alias("s_win_1h"),
    pl.col("Amount_Received_USD")
      .rolling_sum_by("ts", window_size="24h").shift(1).over("From Account")
      .alias("s_win_24h"),
])

# receiver 기준
df_feat = df_feat.sort(["To Account", "ts"])
df_feat = df_feat.with_columns([
    pl.col("Amount_Received_USD")
      .rolling_sum_by("ts", window_size="1h").shift(1).over("To Account")
      .alias("r_win_1h"),
    pl.col("Amount_Received_USD")
      .rolling_sum_by("ts", window_size="24h").shift(1).over("To Account")
      .alias("r_win_24h"),
    pl.col("Amount_Paid_USD")
      .rolling_sum_by("ts", window_size="1h").shift(1).over("To Account")
      .alias("r_wout_1h"),
    pl.col("Amount_Paid_USD")
      .rolling_sum_by("ts", window_size="24h").shift(1).over("To Account")
      .alias("r_wout_24h"),
])

GRAPH_WIN_FEATS = ["s_wout_1h","s_wout_24h","s_win_1h","s_win_24h",
                   "r_win_1h","r_win_24h","r_wout_1h","r_wout_24h"]
df_feat = df_feat.with_columns([
    pl.col(c).fill_null(0).cast(pl.Float32) for c in GRAPH_WIN_FEATS
])
print("5-b 완료:", GRAPH_WIN_FEATS)
```

## 5-c) Edge Pair 피처 (A->B pair 기준 누수방지)

```python
PAIR_KEY = ["From Account", "To Account"]
df_feat = df_feat.sort(PAIR_KEY + ["ts"])

df_feat = df_feat.with_columns([
    pl.col("_one")
      .rolling_sum_by("ts", window_size="1h").shift(1).over(PAIR_KEY)
      .alias("e_cnt_1h_past"),
    pl.col("Amount_Paid_USD")
      .rolling_sum_by("ts", window_size="1h").shift(1).over(PAIR_KEY)
      .alias("e_sum_paid_1h_past"),
    pl.col("_one")
      .rolling_sum_by("ts", window_size="24h").shift(1).over(PAIR_KEY)
      .alias("e_cnt_24h_past"),
    pl.col("Amount_Paid_USD")
      .rolling_sum_by("ts", window_size="24h").shift(1).over(PAIR_KEY)
      .alias("e_sum_paid_24h_past"),
])

train_pairs = (
    df_feat.filter(pl.col("split") == "train")
           .select(PAIR_KEY).unique()
           .with_columns(pl.lit(1).alias("seen_in_train"))
)
df_feat = (
    df_feat.join(train_pairs, on=PAIR_KEY, how="left")
           .with_columns(pl.col("seen_in_train").is_null().alias("e_new_pair_flag"))
           .drop(["seen_in_train"])
)

df_feat = df_feat.with_columns([
    pl.col("e_cnt_1h_past").fill_null(0).cast(pl.Int32),
    pl.col("e_sum_paid_1h_past").fill_null(0).cast(pl.Float32),
    pl.col("e_cnt_24h_past").fill_null(0).cast(pl.Int32),
    pl.col("e_sum_paid_24h_past").fill_null(0).cast(pl.Float32),
    pl.col("e_new_pair_flag").fill_null(False),
])
print("5-c 완료")
```

## 5-d) GroupBy Rolling 피처 (Sender / Receiver past aggregation)

```python
# ============================================================
# 5-d) GroupBy Rolling 피처 (Sender / Receiver)
# ============================================================
df_feat = df_feat.sort(["From Account", "ts"])
df_feat = df_feat.with_columns([
    pl.col("_one").rolling_sum_by("ts","1h").shift(1).over("From Account").alias("s_cnt_1h_past"),
    pl.col("Amount_Paid_USD").rolling_sum_by("ts","1h").shift(1).over("From Account").alias("s_sum_paid_1h_past"),
    pl.col("Amount_Paid_USD").rolling_mean_by("ts","1h").shift(1).over("From Account").alias("s_mean_paid_1h_past"),
    pl.col("Amount_Paid_USD").rolling_max_by("ts","1h").shift(1).over("From Account").alias("s_max_paid_1h_past"),
    pl.col("Amount_Paid_USD").rolling_std_by("ts","1h").shift(1).over("From Account").alias("s_std_paid_1h_past"),
    pl.col("Amount_Received_USD").rolling_sum_by("ts","1h").shift(1).over("From Account").alias("s_sum_recv_1h_past"),
    pl.col("Amount_Received_USD").rolling_mean_by("ts","1h").shift(1).over("From Account").alias("s_mean_recv_1h_past"),
    pl.col("_one").rolling_sum_by("ts","24h").shift(1).over("From Account").alias("s_cnt_24h_past"),
    pl.col("Amount_Paid_USD").rolling_sum_by("ts","24h").shift(1).over("From Account").alias("s_sum_paid_24h_past"),
])
df_feat = df_feat.with_columns(
    (pl.col("s_sum_recv_1h_past") - pl.col("s_sum_paid_1h_past")).alias("s_net_1h_past")
)

df_feat = df_feat.sort(["To Account", "ts"])
df_feat = df_feat.with_columns([
    pl.col("_one").rolling_sum_by("ts","1h").shift(1).over("To Account").alias("r_cnt_1h_past"),
    pl.col("Amount_Received_USD").rolling_sum_by("ts","1h").shift(1).over("To Account").alias("r_sum_recv_1h_past"),
    pl.col("Amount_Received_USD").rolling_mean_by("ts","1h").shift(1).over("To Account").alias("r_mean_recv_1h_past"),
    pl.col("Amount_Received_USD").rolling_max_by("ts","1h").shift(1).over("To Account").alias("r_max_recv_1h_past"),
    pl.col("Amount_Received_USD").rolling_std_by("ts","1h").shift(1).over("To Account").alias("r_std_recv_1h_past"),
    pl.col("Amount_Paid_USD").rolling_sum_by("ts","1h").shift(1).over("To Account").alias("r_sum_paid_1h_past"),
    pl.col("Amount_Paid_USD").rolling_mean_by("ts","1h").shift(1).over("To Account").alias("r_mean_paid_1h_past"),
    pl.col("_one").rolling_sum_by("ts","24h").shift(1).over("To Account").alias("r_cnt_24h_past"),
])
df_feat = df_feat.with_columns(
    (pl.col("r_sum_recv_1h_past") - pl.col("r_sum_paid_1h_past")).alias("r_net_1h_past")
)

PAST_FILL_ZERO = [
    "s_cnt_1h_past","s_sum_paid_1h_past","s_mean_paid_1h_past","s_max_paid_1h_past",
    "s_std_paid_1h_past","s_sum_recv_1h_past","s_mean_recv_1h_past","s_cnt_24h_past",
    "s_sum_paid_24h_past","s_net_1h_past",
    "r_cnt_1h_past","r_sum_recv_1h_past","r_mean_recv_1h_past","r_max_recv_1h_past",
    "r_std_recv_1h_past","r_sum_paid_1h_past","r_mean_paid_1h_past","r_cnt_24h_past",
    "r_net_1h_past",
]
df_feat = df_feat.with_columns([pl.col(c).fill_null(0) for c in PAST_FILL_ZERO])
print("5-d 완료. 총 컬럼:", len(df_feat.columns))
```

# 7) GraphSAGE Stage 1 준비

## 7-a) 노드 피처 구성 함수 (account 집계)

```python
# ── 노드 피처 컬럼 정의 ───────────────────────────────────────────────────────
NODE_FEAT_COLS = [
    "degree_out",          # unique receivers (out-degree)
    "degree_in",           # unique senders   (in-degree)
    "txn_count",           # total transactions
    "unique_counterparty", # degree_out + degree_in
    "sent_recv_ratio",     # log1p(total_sent) / log1p(total_recv)
    "log_mean_paid",       # mean log(Amount_Paid_USD+1)
    "log_max_paid",        # max  log(Amount_Paid_USD+1)
    "log_std_paid",        # std  log(Amount_Paid_USD+1)
    "log_mean_recv",       # mean log(Amount_Received_USD+1)
    "log_max_recv",        # max  log(Amount_Received_USD+1)
    "log_std_recv",        # std  log(Amount_Received_USD+1)
]


def build_node_features(df_pd: pd.DataFrame,
                        scaler=None,
                        fit_scaler: bool = False):
    '''
    account별 노드 피처 + 레이블 생성.
    '''
    sender = df_pd.groupby("From Account", sort=False).agg(
        degree_out=("To Account", "nunique"),
        txn_count_out=("Amount_Paid_USD", "count"),
        total_sent=("Amount_Paid_USD", "sum"),
        log_mean_paid=("log_amount_paid_usd", "mean"),
        log_max_paid=("log_amount_paid_usd", "max"),
        log_std_paid=("log_amount_paid_usd",
                      lambda x: float(x.std(ddof=0)) if len(x) > 1 else 0.0),
    ).reset_index().rename(columns={"From Account": "account"})

    receiver = df_pd.groupby("To Account", sort=False).agg(
        degree_in=("From Account", "nunique"),
        txn_count_in=("Amount_Received_USD", "count"),
        total_recv=("Amount_Received_USD", "sum"),
        log_mean_recv=("log_amount_received_usd", "mean"),
        log_max_recv=("log_amount_received_usd", "max"),
        log_std_recv=("log_amount_received_usd",
                      lambda x: float(x.std(ddof=0)) if len(x) > 1 else 0.0),
    ).reset_index().rename(columns={"To Account": "account"})

    node_feat = sender.merge(receiver, on="account", how="outer").fillna(0.0)

    node_feat["txn_count"]           = node_feat["txn_count_out"]  + node_feat["txn_count_in"]
    node_feat["unique_counterparty"] = node_feat["degree_out"]     + node_feat["degree_in"]
    node_feat["sent_recv_ratio"]     = (
        np.log1p(node_feat["total_sent"]) /
        (np.log1p(node_feat["total_recv"]) + 1e-8)
    )

    from_lbl = (df_pd.groupby("From Account")["Is Laundering"]
                      .max().reset_index()
                      .rename(columns={"From Account": "account", "Is Laundering": "label"}))
    to_lbl   = (df_pd.groupby("To Account")["Is Laundering"]
                      .max().reset_index()
                      .rename(columns={"To Account": "account", "Is Laundering": "label"}))
    node_labels = (pd.concat([from_lbl, to_lbl])
                     .groupby("account")["label"].max()
                     .reset_index())

    feat_arr = node_feat[NODE_FEAT_COLS].values.astype(np.float32)
    if fit_scaler:
        scaler   = RobustScaler()
        feat_arr = np.asarray(scaler.fit_transform(feat_arr), dtype=np.float32)
    elif scaler is not None:
        feat_arr = np.asarray(scaler.transform(feat_arr), dtype=np.float32)

    node_feat_out = node_feat[["account"]].copy()
    for i, col in enumerate(NODE_FEAT_COLS):
        node_feat_out[col] = feat_arr[:, i]

    return node_feat_out, node_labels, scaler


print("build_node_features 함수 정의 완료.")
print(f"노드 피처 {len(NODE_FEAT_COLS)}개: {NODE_FEAT_COLS}")
```

## 7-b) PyG Data 구성 함수

```python
def build_single_graph(all_df, node_feat_df, node_label_df):
    '''
    단일 그래프 생성 (1개 + 마스크 방식 - 메모리 1/3).
    '''
    accounts = node_feat_df["account"].tolist()
    acc2idx  = {acc: i for i, acc in enumerate(accounts)}
    N        = len(accounts)

    mask = (
        all_df["From Account"].isin(acc2idx) &
        all_df["To Account"].isin(acc2idx)
    )
    t   = all_df[mask]
    src = t["From Account"].map(acc2idx).values
    dst = t["To Account"].map(acc2idx).values
    edge_index = torch.tensor(np.stack([src, dst], axis=0), dtype=torch.long)

    x = torch.tensor(node_feat_df[NODE_FEAT_COLS].values, dtype=torch.float32)

    print("NaN in features:", torch.isnan(x).sum())
    print("Inf in features:", torch.isinf(x).sum())

    x = torch.nan_to_num(x, nan=0.0, posinf=0.0, neginf=0.0)
    lbl_map = dict(zip(node_label_df["account"], node_label_df["label"]))
    y       = torch.tensor(
        [lbl_map.get(acc, 0) for acc in accounts], dtype=torch.float32)

    data = Data(x=x, edge_index=edge_index, y=y)
    print(f"  Nodes={N:,}  Edges={data.num_edges:,}  "
          f"Pos={int(y.sum()):,} ({100*float(y.mean()):.2f}%)")
    return data, acc2idx


def make_split_mask(acc2idx, split_df):
    '''split_df에 등장하는 계좌 -> bool 마스크'''
    accs = set(split_df["From Account"].tolist() + split_df["To Account"].tolist())
    mask = torch.zeros(len(acc2idx), dtype=torch.bool)
    for acc in accs:
        if acc in acc2idx:
            mask[acc2idx[acc]] = True
    return mask


print("build_single_graph / make_split_mask 정의 완료.")
```

## 7-c) GraphSAGE 모델 (2-layer)

```python
class TwoLayerSAGE(nn.Module):
    '''
    2-layer GraphSAGE for binary node classification.
    aggr1: conv1 집계 방식 ("mean" | "max")  ← pool = max
    aggr2: conv2 집계 방식 ("mean" | "max")
    조합 예시:
      mean_mean  → aggr1="mean", aggr2="mean"
      mean_pool  → aggr1="mean", aggr2="max"
      pool_pool  → aggr1="max",  aggr2="max"
    출력: (N,) logits -> sigmoid -> probability
    '''
    def __init__(self, in_channels: int, hidden_channels: int,
                 dropout: float = 0.5,
                 aggr1: str = "max", aggr2: str = "max"):
        super().__init__()
        self.conv1   = SAGEConv(in_channels, hidden_channels, aggr=aggr1)
        self.conv2   = SAGEConv(hidden_channels, 1,            aggr=aggr2)
        self.dropout = dropout

    def forward(self, x, edge_index):
        x = self.conv1(x, edge_index)
        x = F.relu(x)
        x = F.dropout(x, p=self.dropout, training=self.training)
        x = self.conv2(x, edge_index)
        return x.squeeze(-1)          # (N,) logits

    def embed(self, x, edge_index):
        '''중간 임베딩 반환 (시각화용)'''
        x = self.conv1(x, edge_index)
        x = F.relu(x)
        return x                      # (N, hidden_channels)


# aggr 조합 매핑 (sweep config 값 → (aggr1, aggr2))
AGGR_COMBOS = {
    "mean_mean": ("mean", "mean"),
    "mean_pool": ("mean", "max"),
    "pool_pool": ("max",  "max"),
}

print("TwoLayerSAGE 모델 정의 완료. (aggr1/aggr2 파라미터화)")
```

```python
# ── GraphSAGE 학습 / 추론 (NeighborLoader 기반 — 대규모 그래프 호환) ──────────
# 이전 k_hop_subgraph 방식은 노드 200만+ 규모에서 RAM 폭발 → NeighborLoader로 교체
from torch_geometric.loader import NeighborLoader
import gc

def train_graphsage_epoch(model, data, train_mask, optimizer, pos_weight,
                          num_neighbors=(15, 10), batch_size=1024):
    """
    NeighborLoader 기반 mini-batch 학습.
    num_neighbors: 각 hop에서 샘플링할 이웃 수 (layer당)
                   (15, 10) → 1-hop 15개, 2-hop 10개 → 배치당 최대 ~150 노드
    """
    model.train()
    criterion = nn.BCEWithLogitsLoss(
        pos_weight=torch.tensor([pos_weight], device=DEVICE)
    )

    loader = NeighborLoader(
        data,
        num_neighbors=list(num_neighbors),
        batch_size=batch_size,
        input_nodes=train_mask,
        shuffle=True,
        num_workers=0,
        subgraph_type="induced",  # pyg-lib/torch-sparse 없이도 동작
    )

    total_loss, n = 0.0, 0
    for batch in loader:
        batch = batch.to(DEVICE)
        # batch.batch_size: 현재 배치의 seed 노드 수
        out  = model(batch.x.float(), batch.edge_index)[:batch.batch_size]
        tgt  = batch.y[:batch.batch_size].float()
        loss = criterion(out, tgt)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
        n += 1

    return total_loss / max(n, 1)


@torch.no_grad()
def infer_graphsage(model, data, batch_size=2048, num_neighbors=(15, 10)):
    """
    NeighborLoader 기반 전체 노드 추론.
    모든 노드를 seed로 삼아 순서대로 처리 → probs 리스트 반환.
    """

    model = model.to(DEVICE)
    model.eval()
    num_nodes = data.num_nodes

    all_mask = torch.ones(num_nodes, dtype=torch.bool)

    loader = NeighborLoader(
        data,
        num_neighbors=list(num_neighbors),
        batch_size=batch_size,
        input_nodes=all_mask,
        shuffle=False,
        num_workers=0,
        subgraph_type="induced",  # pyg-lib/torch-sparse 없이도 동작
    )

    probs = torch.zeros(num_nodes)
    for batch in loader:
        batch_size_actual = batch.batch_size
       # 먼저 GPU로 이동
        batch = batch.to(DEVICE)
       # seed index는 CPU로 가져오기
        seed_idx = batch.n_id[:batch_size_actual].cpu()
        out = model(batch.x.float(), batch.edge_index)[:batch_size_actual]
        p = torch.sigmoid(out).cpu()
        probs[seed_idx] = p

    # for batch in loader:
    #     batch_size_actual = batch.batch_size
    #     # input_id: seed 노드들의 원래 global index
    #     seed_idx = batch.n_id[:batch_size_actual]
    #     batch = batch.to(DEVICE)
    #     out   = model(batch.x.float(), batch.edge_index)[:batch_size_actual]
    #     p     = torch.sigmoid(out).cpu()
    #     probs[seed_idx] = p


        del batch, out, p

    torch.cuda.empty_cache()

    print("NaN in probs:", torch.isnan(probs).sum())
    print("Total probs:", len(probs))
    probs = torch.nan_to_num(probs)
    return probs.tolist()



def attach_sage_probs(df, acc2idx, sage_probs, prefix="sage"):
    df = df.copy()
    df[f"{prefix}_prob_sender"] = np.array(
        [sage_probs[acc2idx[a]] if a in acc2idx else 0.0
         for a in df["From Account"]], dtype=np.float32
    )
    df[f"{prefix}_prob_receiver"] = np.array(
        [sage_probs[acc2idx[a]] if a in acc2idx else 0.0
         for a in df["To Account"]], dtype=np.float32
    )
    return df

print("GraphSAGE 유틸 함수 정의 완료 (NeighborLoader 기반, 대규모 그래프 호환).")


import numpy as np

```

# 8) 평가 유틸 (W&B 로깅)

```python
def safe_auc(y_true, y_score):
    if len(np.unique(y_true)) <= 1:
        return np.nan, np.nan
    return roc_auc_score(y_true, y_score), average_precision_score(y_true, y_score)


def workload_to_find_k_positives(y_true, y_score, k_pos):
    """
    목표: true positive를 k_pos개 '찾기' 위해
         상위 점수부터 몇 건(N)을 조사해야 하는지 계산
    Returns:
      N_required, precision, recall, f1, k_found
    """
    y_true  = np.asarray(y_true).astype(int)
    y_score = np.asarray(y_score)
    total_pos = y_true.sum()
    if total_pos == 0:
        return 0, 0.0, 0.0, 0.0, 0
    order  = np.argsort(-y_score)
    cum_tp = np.cumsum(y_true[order])
    target = min(k_pos, int(total_pos))
    idx    = int(np.searchsorted(cum_tp, target, side="left"))
    N_req  = idx + 1
    k_found = int(cum_tp[idx])
    prec = k_found / (N_req + 1e-12)
    rec  = k_found / (total_pos + 1e-12)
    fp   = N_req - k_found
    fn   = total_pos - k_found
    f1   = (2 * k_found) / (2 * k_found + fp + fn + 1e-12)
    return N_req, prec, rec, f1, k_found


def threshold_at_recall(y_true, y_score, desired_recall=0.90):
    prec, rec, thr = precision_recall_curve(y_true, y_score)
    if len(thr) == 0:
        return 0.5, prec, rec, thr
    i = int(np.argmin(np.abs(rec[1:] - desired_recall)))
    return float(thr[i]), prec, rec, thr


def score_bin_distribution(y_true, y_score, n_bins=10, score_scale=1000):
    y = np.asarray(y_true).astype(int)
    s = np.clip(np.asarray(y_score) * score_scale, 0, score_scale)
    bins    = np.linspace(0, score_scale, n_bins + 1)
    bin_idx = np.clip(np.digitize(s, bins, right=False) - 1, 0, n_bins - 1)
    df = pd.DataFrame({"bin": bin_idx, "label": y})
    cnt = (df.groupby(["bin", "label"]).size().unstack(fill_value=0)
             .reindex(columns=[0, 1], fill_value=0)
             .rename(columns={0: "normal_cnt", 1: "laundering_cnt"})
             .reset_index())
    cnt["bin_range"] = cnt["bin"].apply(lambda i: f"{int(bins[i])}-{int(bins[i+1])}")
    return cnt[["bin", "bin_range", "normal_cnt", "laundering_cnt"]]


def aggregate_to_account_level(df_meta, y_true, y_score,
                                account_cols=("From Account",), score_agg="max"):
    """거래 단위 (y_true, y_score)를 계좌 단위로 집계."""
    y, s = np.asarray(y_true).astype(int), np.asarray(y_score)
    tmp  = df_meta[list(account_cols)].copy()
    tmp["label_trx"] = y; tmp["score_trx"] = s
    if len(account_cols) == 2:
        a1, a2 = account_cols
        long = pd.concat([
            tmp[[a1,"label_trx","score_trx"]].rename(columns={a1:"account"}),
            tmp[[a2,"label_trx","score_trx"]].rename(columns={a2:"account"}),
        ], ignore_index=True)
    else:
        long = tmp[[account_cols[0],"label_trx","score_trx"]].rename(
            columns={account_cols[0]:"account"})
    long = long.dropna(subset=["account"])
    agg_fn = "mean" if score_agg == "mean" else "max"
    return (long.groupby("account")
                .agg(label=("label_trx","max"), score=("score_trx", agg_fn),
                     n_hits=("score_trx","size"))
                .reset_index())


print("평가 유틸 함수 정의 완료.")
```

# 8-2) Top-k 로직 (하루 기준/누적 기준)

```python
# ── Per-day workload summary ──────────────────────────────────────────────────
def per_day_workload_summary(df, day_col, label_col, score_col, k_pos_list=(30,50,100,200)):
    """
    day별로 'k_pos개 적발하기 위해 필요한 조사량 N' 계산 후
    k_pos마다 daily 분포 + mean/median/p90 요약을 반환
    """
    out_rows = []
    for day, g in df.groupby(day_col):
        y = g[label_col].astype(int).values
        s = g[score_col].values
        total_pos = int(y.sum())
        n = len(g)
        for k_pos in k_pos_list:
            N_req, p, r, f1, found = workload_to_find_k_positives(y, s, k_pos)
            out_rows.append({
                "day": day, "n_rows": n, "pos": total_pos, "k_pos": k_pos,
                "N_required": N_req, "precision_at_Nk": p, "recall_at_Nk": r,
                "f1_at_Nk": f1, "found_tp": found,
                "target_k": min(k_pos, total_pos),
                "hit_target": (found >= min(k_pos, total_pos)) and (total_pos > 0)
            })
    daily = pd.DataFrame(out_rows)
    daily_pos = daily[daily["pos"] > 0].copy()

    def summarize(sub):
        return pd.Series({
            "days": sub["day"].nunique(),
            "mean_N_required": sub["N_required"].mean(),
            "median_N_required": sub["N_required"].median(),
            "p90_N_required": sub["N_required"].quantile(0.90),
            "mean_precision": sub["precision_at_Nk"].mean(),
            "median_precision": sub["precision_at_Nk"].median(),
            "p90_precision": sub["precision_at_Nk"].quantile(0.90),
        })

    summary = (
        daily_pos.groupby("k_pos", as_index=False)
        .apply(summarize)
        .reset_index(drop=True)
    )
    return daily, summary


def log_per_day_kpi_to_wandb(split_name, df_raw_with_day, y_true, y_score,
                              k_list=(30,50,100,200)):
    tmp = pd.DataFrame({
        "ts_day": df_raw_with_day["ts_day"].values,
        "label": np.asarray(y_true).astype(int),
        "score": np.asarray(y_score),
    })
    daily, summary = per_day_workload_summary(
        tmp, day_col="ts_day", label_col="label", score_col="score", k_pos_list=k_list
    )
    for _, row in summary.iterrows():
        k = int(row["k_pos"])
        wandb.log({
            f"{split_name}_perday_find_{k}_mean_N": float(row["mean_N_required"]),
            f"{split_name}_perday_find_{k}_median_N": float(row["median_N_required"]),
            f"{split_name}_perday_find_{k}_p90_N": float(row["p90_N_required"]),
            f"{split_name}_perday_find_{k}_mean_precision": float(row["mean_precision"]),
            f"{split_name}_perday_find_{k}_median_precision": float(row["median_precision"]),
            f"{split_name}_perday_find_{k}_p90_precision": float(row["p90_precision"]),
            f"{split_name}_perday_days_with_pos_{k}": int(row["days"]),
        })
    table = wandb.Table(dataframe=daily.head(5000))
    wandb.log({f"{split_name}_perday_workload_table": table})


def topN_distinct_account_metrics(df_meta, y_true, y_score, topN,
                                   account_cols=("From Account","To Account"),
                                   score_agg="max"):
    y = np.asarray(y_true).astype(int)
    s = np.asarray(y_score)
    tmp = df_meta[list(account_cols)].copy()
    tmp["label_trx"] = y; tmp["score_trx"] = s
    tmp = tmp.sort_values("score_trx", ascending=False).head(int(min(topN, len(tmp))))
    agg = aggregate_to_account_level(
        df_meta=tmp, y_true=tmp["label_trx"].values, y_score=tmp["score_trx"].values,
        account_cols=account_cols, score_agg=score_agg
    )
    y_acc = agg["label"].values
    tp = int((y_acc == 1).sum())
    fp = int((y_acc == 0).sum())
    precision = tp / (tp + fp + 1e-12)
    return {
        "topN_trx": int(min(topN, len(tmp))),
        "distinct_accounts": int(len(agg)),
        "tp_accounts": tp, "fp_accounts": fp,
        "precision_accounts": float(precision),
    }


def log_per_day_account_kpi_to_wandb(split_name, df_meta, y_true, y_score,
                                      k_list=(30,50,100,200),
                                      account_cols=("From Account","To Account"),
                                      score_agg="max"):
    y = np.asarray(y_true).astype(int)
    s = np.asarray(y_score)
    tmp = df_meta[["ts_day"] + list(account_cols)].copy()
    tmp["label_trx"] = y; tmp["score_trx"] = s
    rows = []
    for day, g in tmp.groupby("ts_day"):
        agg = aggregate_to_account_level(
            df_meta=g, y_true=g["label_trx"].values, y_score=g["score_trx"].values,
            account_cols=account_cols, score_agg=score_agg
        )
        y_acc = agg["label"].values; s_acc = agg["score"].values
        for k_pos in k_list:
            N_req, p, r, f1, found = workload_to_find_k_positives(y_acc, s_acc, k_pos)
            rows.append({"ts_day": day, "k_pos": k_pos,
                         "n_accounts": len(agg), "pos_accounts": int(y_acc.sum()),
                         "N_required": N_req, "precision": p, "recall": r,
                         "f1": f1, "found_tp": found})
    daily = pd.DataFrame(rows)
    daily_pos = daily[daily["pos_accounts"] > 0].copy()
    if len(daily_pos) > 0:
        summ = daily_pos.groupby("k_pos").agg(
            days=("ts_day","nunique"), mean_N=("N_required","mean"),
            median_N=("N_required","median"),
            p90_N=("N_required", lambda x: x.quantile(0.9)),
            mean_precision=("precision","mean"),
        ).reset_index()
        for _, r in summ.iterrows():
            k = int(r["k_pos"])
            wandb.log({
                f"{split_name}_acc_perday_find_{k}_mean_N": float(r["mean_N"]),
                f"{split_name}_acc_perday_find_{k}_median_N": float(r["median_N"]),
                f"{split_name}_acc_perday_find_{k}_p90_N": float(r["p90_N"]),
                f"{split_name}_acc_perday_find_{k}_mean_precision": float(r["mean_precision"]),
                f"{split_name}_acc_perday_days_with_pos_{k}": int(r["days"]),
            })
    wandb.log({f"{split_name}_acc_perday_table": wandb.Table(dataframe=daily.head(5000))})


print("Top-k 로직 (하루/누적 기준) 정의 완료.")
```

```python
# ── AML 패턴 분석 함수 ────────────────────────────────────────────────────────
def get_pattern_analysis_tables(df, prob, thr):
    """패턴별/상세별 Recall을 계산하여 WandB Table 2개를 반환"""
    analysis = pd.DataFrame({
        "Pattern_Type":   df["Pattern_Type"],
        "Pattern_Detail": df["Pattern_Detail"],
        "y_true":         df["Is Laundering"],
        "is_detected":    (prob >= thr).astype(int)
    })
    pattern_df = analysis[analysis["y_true"] == 1].copy()
    if pattern_df.empty:
        return (wandb.Table(columns=["Pattern_Type","Recall_Pct"]),
                wandb.Table(columns=["Pattern_Type","Pattern_Detail","Recall_Pct"]))
    summary = pattern_df.groupby("Pattern_Type").agg(
        Total_Actual=("is_detected","count"),
        Detected_Count=("is_detected","sum")
    ).reset_index()
    summary["Recall_Pct"] = (summary["Detected_Count"] / summary["Total_Actual"] * 100).round(2)
    summary = summary.sort_values("Recall_Pct")
    detail = pattern_df[pattern_df["Pattern_Detail"] != ""].groupby(
        ["Pattern_Type","Pattern_Detail"]).agg(
        Total_Actual=("is_detected","count"),
        Detected_Count=("is_detected","sum")
    ).reset_index()
    if not detail.empty:
        detail["Recall_Pct"] = (detail["Detected_Count"] / detail["Total_Actual"] * 100).round(2)
        detail = detail.sort_values(["Pattern_Type","Recall_Pct"])
    return wandb.Table(dataframe=summary), wandb.Table(dataframe=detail)


def get_pattern_analysis_tables_find_k(df, prob, k_pos_target):
    """Find-K 로직 기반으로 패턴별/상세별 Recall을 계산."""
    y_true  = df["Is Laundering"].astype(int).values
    y_score = np.asarray(prob)
    N_req, _, _, _, _ = workload_to_find_k_positives(y_true, y_score, k_pos_target)
    analysis = pd.DataFrame({
        "Pattern_Type":   df["Pattern_Type"],
        "Pattern_Detail": df["Pattern_Detail"],
        "y_true":         y_true,
        "score":          y_score
    })
    analysis = analysis.sort_values("score", ascending=False).reset_index(drop=True)
    analysis["is_detected"] = 0
    analysis.loc[:N_req-1, "is_detected"] = 1
    pattern_df = analysis[analysis["y_true"] == 1].copy()
    if pattern_df.empty:
        return (wandb.Table(columns=["Pattern_Type","Total_Actual","Detected_Count","Recall_Pct"]),
                wandb.Table(columns=["Pattern_Type","Pattern_Detail","Total_Actual","Detected_Count","Recall_Pct"]))
    summary = pattern_df.groupby("Pattern_Type").agg(
        Total_Actual=("is_detected","count"),
        Detected_Count=("is_detected","sum")
    ).reset_index()
    summary["Recall_Pct"] = (summary["Detected_Count"] / summary["Total_Actual"] * 100).round(2)
    summary = summary.sort_values("Recall_Pct")
    detail = pattern_df[pattern_df["Pattern_Detail"] != ""].groupby(
        ["Pattern_Type","Pattern_Detail"]).agg(
        Total_Actual=("is_detected","count"),
        Detected_Count=("is_detected","sum")
    ).reset_index()
    if not detail.empty:
        detail["Recall_Pct"] = (detail["Detected_Count"] / detail["Total_Actual"] * 100).round(2)
        detail = detail.sort_values(["Pattern_Type","Recall_Pct"])
    return wandb.Table(dataframe=summary), wandb.Table(dataframe=detail)


def log_find_k_error_analysis(split_name, df, y_true, y_score, k_pos_target, top_features):
    """Find-K 로직 기반 에러 분석 (TP/FP/FN 그룹별 분석)."""
    y_true_arr  = np.asarray(y_true).astype(int)
    y_score_arr = np.asarray(y_score)
    N_req, _, _, _, _ = workload_to_find_k_positives(y_true_arr, y_score_arr, k_pos_target)
    analysis_df = df.copy()
    analysis_df["prob"]  = y_score_arr
    analysis_df["label"] = y_true_arr
    analysis_df = analysis_df.sort_values("prob", ascending=False).reset_index(drop=True)
    analysis_df["pred"] = 0
    analysis_df.loc[:N_req-1, "pred"] = 1
    def categorize(row):
        if row["label"] == 1 and row["pred"] == 1: return "TP (정탐)"
        if row["label"] == 0 and row["pred"] == 1: return "FP (오탐)"
        if row["label"] == 1 and row["pred"] == 0: return "FN (미탐)"
        return "TN"
    analysis_df["error_cat"] = analysis_df.apply(categorize, axis=1)
    error_by_day = analysis_df.groupby(["ts_day","error_cat"]).size().unstack(fill_value=0).reset_index()
    wandb.log({f"{split_name}_find_{k_pos_target}_daily_error_trend": wandb.Table(dataframe=error_by_day)})
    feat_stats = []
    for cat in ["TP (정탐)", "FP (오탐)", "FN (미탐)"]:
        sub = analysis_df[analysis_df["error_cat"] == cat]
        if sub.empty: continue
        stat_row = {"category": cat, "count": len(sub)}
        for f in top_features:
            if f not in sub.columns: continue
            val = sub[f]
            if pd.api.types.is_numeric_dtype(val):
                stat_row[f"{f}_mean"] = round(val.mean(), 4)
        feat_stats.append(stat_row)
    wandb.log({f"{split_name}_find_{k_pos_target}_feat_stats": wandb.Table(dataframe=pd.DataFrame(feat_stats))})
    fn_df = analysis_df[analysis_df["error_cat"] == "FN (미탐)"]
    if not fn_df.empty and "Pattern_Type" in fn_df.columns:
        fn_pat = fn_df.groupby(["ts_day","Pattern_Type"]).size().reset_index(name="fn_count")
        wandb.log({f"{split_name}_find_{k_pos_target}_missed_patterns_by_day": wandb.Table(dataframe=fn_pat)})


print("패턴 분석 / 에러 분석 유틸 함수 정의 완료.")


# ── Score 분포 bin 테이블 ──────────────────────────────────────────────────────
def score_bin_distribution(y_true, y_score, n_bins=10, score_scale=1000):
    """
    score를 0~score_scale로 스케일링한 뒤,
    bin별 label(0/1) 건수 집계.
    항상 normal_cnt / laundering_cnt 컬럼이 존재하도록 보정.
    """
    y = np.asarray(y_true).astype(int)
    s = np.asarray(y_score)

    s_scaled = np.clip(s * score_scale, 0, score_scale)
    bins = np.linspace(0, score_scale, n_bins + 1)

    # 0..n_bins-1
    bin_idx = np.digitize(s_scaled, bins, right=False) - 1
    bin_idx = np.clip(bin_idx, 0, n_bins - 1)

    df = pd.DataFrame({"bin": bin_idx, "label": y})

    # label별 count (항상 0/1 둘 다 컬럼을 만들도록 reindex)
    cnt = (
        df.groupby(["bin", "label"])
          .size()
          .unstack(fill_value=0)
          .reindex(columns=[0, 1], fill_value=0)   # ✅ 핵심
          .rename(columns={0: "normal_cnt", 1: "laundering_cnt"})
          .reset_index()
    )

    # bin_range 붙이기
    cnt["bin_range"] = cnt["bin"].apply(lambda i: f"{int(bins[i])}-{int(bins[i+1])}")

    # 보기 좋게 정렬
    cnt = cnt[["bin", "bin_range", "normal_cnt", "laundering_cnt"]]
    return cnt


# ── TP/FP/FN 피처 분포 + 일자별 추이 분석 ──────────────────────────────────────
def log_detailed_error_analysis(split_name, df, y_true, y_score, threshold, top_features):
    """
    정탐(TP), 오탐(FP), 미탐(FN) 그룹별로 피처 분포와 일자별 추이를 분석하여 WandB에 로깅
    """
    analysis_df = df.copy()
    analysis_df['prob'] = y_score
    analysis_df['pred'] = (y_score >= threshold).astype(int)
    analysis_df['label'] = y_true.astype(int)

    # TP, FP, FN, TN 분류
    def categorize(row):
        if row['label'] == 1 and row['pred'] == 1: return 'TP (정탐)'
        if row['label'] == 0 and row['pred'] == 1: return 'FP (오탐)'
        if row['label'] == 1 and row['pred'] == 0: return 'FN (미탐)'
        return 'TN'

    analysis_df['error_cat'] = analysis_df.apply(categorize, axis=1)

    # (1) 일자별 에러 분포 (정탐/오탐/미탐 건수 추이)
    # 2022/09/14 ~ 2022/09/28 범위의 트렌드 파악용
    error_by_day = analysis_df.groupby(['ts_day', 'error_cat']).size().unstack(fill_value=0).reset_index()

    thr_tag = f"thr{threshold:.2f}"

    wandb.log({f"{split_name}_{thr_tag}_daily_error_trend": wandb.Table(dataframe=error_by_day)})

    # (2) 피처별 에러 특성 분석 (어떤 피처가 오탐/미탐에 영향을 주는가?)
    # 중요도 상위 피처(top_features)에 대해 각 그룹별 평균 또는 빈도 계산
    feat_stats = []
    for cat in ['TP (정탐)', 'FP (오탐)', 'FN (미탐)']:
        sub = analysis_df[analysis_df['error_cat'] == cat]
        if sub.empty: continue

        stat_row = {'category': cat, 'count': len(sub)}
        for f in top_features:
            if f not in sub.columns: continue
            val = sub[f]
            # 수치형 피처: 평균값 비교
            if pd.api.types.is_numeric_dtype(val):
                stat_row[f"{f}_mean"] = round(val.mean(), 4)
            # 범주형 피처: 최빈값(Mode) 비교
            else:
                stat_row[f"{f}_top_mode"] = str(val.mode()[0]) if not val.mode().empty else "N/A"
        feat_stats.append(stat_row)

    wandb.log({f"{split_name}_{thr_tag}_feature_characteristics_by_error": wandb.Table(dataframe=pd.DataFrame(feat_stats))})


# ── FN/FP 패턴 중심 에러 분석 ───────────────────────────────────────────────
def log_pattern_error_analysis(split_name, df, y_true, y_score, threshold):
    """
    미탐(FN)과 오탐(FP) 건들에 대해 일자별 패턴 분포를 집중 분석
    """
    df_err = df.copy()
    df_err['pred'] = (y_score >= threshold).astype(int)
    df_err['label'] = y_true.astype(int)

    # FN (미탐): 실제 세탁인데 놓친 것
    fn_df = df_err[(df_err['label'] == 1) & (df_err['pred'] == 0)]
    # FP (오탐): 정상인데 세탁으로 예측한 것
    fp_df = df_err[(df_err['label'] == 0) & (df_err['pred'] == 1)]

    thr_tag = f"thr{threshold:.2f}"

    # (1) 미탐(FN) 패턴 분석: 어떤 날에 어떤 패턴을 주로 놓쳤나?
    if not fn_df.empty:
        fn_pattern_day = fn_df.groupby(['ts_day', 'Pattern_Type']).size().reset_index(name='missed_count')
        wandb.log({f"{split_name}_{thr_tag}_missed_patterns_by_day": wandb.Table(dataframe=fn_pattern_day)})

        # 상세 패턴 미탐 (Pattern_Detail)
        fn_detail = fn_df.groupby(['Pattern_Type', 'Pattern_Detail']).size().reset_index(name='fn_count')
        wandb.log({f"{split_name}_{thr_tag}_missed_patterns_detail": wandb.Table(dataframe=fn_detail)})

    # (2) 오탐(FP) 패턴 분석: 오탐된 건들은 주로 어떤 패턴(혹은 일반 거래)인가?
    if not fp_df.empty:
        # 정상 거래에 패턴 정보가 태깅되어 있는 경우(Synthetic 데이터 등) 유용함
        fp_pattern = fp_df.groupby(['ts_day', 'Pattern_Type']).size().reset_index(name='false_alarm_count')
        wandb.log({f"{split_name}_{thr_tag}_false_alarm_patterns_by_day": wandb.Table(dataframe=fp_pattern)})
```

# 9) Joint Sweep 함수 (GraphSAGE -> XGB)

```python
# ── Node-level AUC (split_df 기반) ──────────────────────────────────────────
def node_auc_split(acc2idx, sage_probs, split_df, name):
    lbl_map = {}
    for acct_col in ["From Account", "To Account"]:
        for acc, lbl in zip(split_df[acct_col], split_df["Is Laundering"]):
            lbl_map[acc] = max(lbl_map.get(acc, 0), lbl)

    accs, labels, probs_list = [], [], []
    for acc, lbl in lbl_map.items():
        if acc in acc2idx:
            accs.append(acc)
            labels.append(lbl)
            probs_list.append(sage_probs[acc2idx[acc]])

    labels    = np.array(labels, dtype=int)
    probs_arr = np.array(probs_list, dtype=np.float32)
    try:
        roc, ap = safe_auc(labels, probs_arr)
        wandb.log({f"sage_{name}_node_roc": roc, f"sage_{name}_node_auprc": ap})
        print(f"  [SAGE] {name} node ROC={roc:.4f}  AUPRC={ap:.4f}")
    except Exception as e:
        print(f"  [SAGE] {name} node AUC error: {e}")


def train_eval_sweep():
    '''
    W&B agent가 호출하는 joint sweep 함수.
    Stage1: GraphSAGE (단일 그래프 + 마스크 방식)
    Stage2: XGBoost  (edge/transaction classification)
    META_COLS는 학습 x, 평가 시에만 사용.
    '''
    run    = wandb.init()
    config = wandb.config
    set_config(transform_output="pandas")
    import gc

    NODE_REQUIRED_COLS = [
        "From Account", "To Account",
        "Amount_Paid_USD", "Amount_Received_USD",
        "log_amount_paid_usd", "log_amount_received_usd",
        "Is Laundering",
    ]

    # ====================================================================
    # ① GraphSAGE Stage 1 (단일 그래프 + 마스크)
    #    LOAD_SAGE_CACHE=True 이면 캐시에서 로드 → SAGE 재학습 스킵 (시간 절약)
    # ====================================================================
    aggr_combo = getattr(config, "sage_aggr_combo", "max_max")
    aggr1, aggr2 = AGGR_COMBOS.get(aggr_combo, ("max", "max"))

    # ── 캐시 사용 여부 결정
    _use_cache = (
        "LOAD_SAGE_CACHE" in dir() and LOAD_SAGE_CACHE
        and sage_probs_cache is not None
        and acc2idx_cache is not None
    )

    if _use_cache:
        print(f"[SAGE] ✅ 캐시 로드 — SAGE 재학습 스킵 (aggr 로그만 기록)")
        sage_probs = sage_probs_cache
        acc2idx    = acc2idx_cache
        node_scaler = node_scaler_cache
        model      = None   # 캐시 모드에서는 model 없음
        wandb.log({"sage_aggr_combo": aggr_combo + "_CACHED",
                   "sage_aggr1": aggr1, "sage_aggr2": aggr2})
    else:
        print("[SAGE] Building node features...")
        _, _, node_scaler = build_node_features(train_df, fit_scaler=True)

        all_slim = pd.concat([
            train_df[NODE_REQUIRED_COLS],
            val_df[NODE_REQUIRED_COLS],
            test_df[NODE_REQUIRED_COLS],
        ], ignore_index=True)

        all_nf, all_nl, _ = build_node_features(
            all_slim, scaler=node_scaler, fit_scaler=False)
        gc.collect()

        print("[SAGE] Building single graph...")
        data, acc2idx = build_single_graph(all_slim, all_nf, all_nl)
        del all_slim, all_nf
        gc.collect()

        train_mask = make_split_mask(acc2idx, train_df)
        print(f"  train_mask: {int(train_mask.sum()):,} nodes")

        sage_hidden  = config.sage_hidden_channels
        sage_dropout = config.sage_dropout
        sage_lr      = config.sage_lr
        sage_epochs  = config.sage_epochs
        n_pos        = int(data.y[train_mask].sum())
        n_neg        = int(train_mask.sum()) - n_pos
        sage_pos_w   = config.sage_pos_weight if hasattr(config, "sage_pos_weight")                        else n_neg / max(n_pos, 1)

        print(f"[SAGE] pos_weight={sage_pos_w:.1f}  hidden={sage_hidden}  dropout={sage_dropout}  aggr={aggr_combo}")

        model     = TwoLayerSAGE(len(NODE_FEAT_COLS), sage_hidden, sage_dropout,
                                  aggr1=aggr1, aggr2=aggr2).to(DEVICE)
        optimizer = torch.optim.Adam(model.parameters(), lr=sage_lr, weight_decay=1e-5)

        best_state, best_loss, patience_cnt = None, float("inf"), 0
        PATIENCE = max(3, sage_epochs // 10)

        for epoch in range(1, sage_epochs + 1):
            loss = train_graphsage_epoch(model, data, train_mask, optimizer, sage_pos_w)
            wandb.log({"sage_train_loss": loss, "sage_epoch": epoch})
            if epoch % 5 == 0:
                print(f"  epoch {epoch:3d}  loss={loss:.4f}")
            if loss < best_loss - 1e-5:
                best_loss    = loss
                best_state   = {k: v.clone() for k, v in model.state_dict().items()}
                patience_cnt = 0
            else:
                patience_cnt += 1
            if patience_cnt >= PATIENCE:
                print(f"  Early stop @ epoch {epoch}")
                break

        if best_state:
            model.load_state_dict(best_state)
        model = model.cpu()

        print("[SAGE] Inferring (single pass)...")
        sage_probs = infer_graphsage(model, data)
        del data
        gc.collect()

        wandb.log({"sage_aggr_combo": aggr_combo, "sage_aggr1": aggr1, "sage_aggr2": aggr2})

    node_auc_split(acc2idx, sage_probs, train_df, "train")
    node_auc_split(acc2idx, sage_probs, val_df,   "val")
    node_auc_split(acc2idx, sage_probs, test_df,  "test")

    train_df_aug = attach_sage_probs(train_df, acc2idx, sage_probs)
    val_df_aug   = attach_sage_probs(val_df,   acc2idx, sage_probs)
    test_df_aug  = attach_sage_probs(test_df,  acc2idx, sage_probs)
    del sage_probs, all_nl
    gc.collect()

    print("[SAGE] 완료.")

    # ====================================================================
    # ② XGBoost Stage 2
    # ====================================================================
    # META_COLS는 학습 x, 평가 시에만 사용
    DROP_META = [LABEL_COL] + META_COLS

    X_train_raw = train_df_aug.drop(columns=[c for c in DROP_META if c in train_df_aug.columns])
    y_train     = train_df_aug[LABEL_COL].astype(int)
    X_val_raw   = val_df_aug.drop(columns=[c for c in DROP_META if c in val_df_aug.columns])
    y_val       = val_df_aug[LABEL_COL].astype(int)
    X_test_raw  = test_df_aug.drop(columns=[c for c in DROP_META if c in test_df_aug.columns])
    y_test      = test_df_aug[LABEL_COL].astype(int)

    # ── 범주형 컬럼 처리 ──────────────────────────────────────────────────
    NATIVE_CAT_COLS = ["dow_cat","Receiving Currency","Payment Currency",
                       "Payment Format","timeofday_bucket",
                       "Sender_Bank_Branch_ID","Receiver_Bank_Branch_ID"]
    HIGH_CARD_COLS  = ["Sender_Entity","Sender_Bank_Name",
                       "Receiver_Entity","Receiver_Bank_Name",
                       "From Bank","To Bank","From Country","To Country"]

    def fit_topk_maps(df_tr, cols, top_k=2000):
        maps = {}
        for c in cols:
            if c not in df_tr.columns: continue
            s  = df_tr[c].astype("string").fillna("__MISSING__")
            vc = s.value_counts(dropna=False)
            maps[c] = set(vc.head(top_k).index.astype(str)) | {"__MISSING__"}
        return maps

    def apply_topk_maps(df, maps, rare="__RARE__"):
        df = df.copy()
        for c, allowed in maps.items():
            if c not in df.columns: continue
            s = df[c].astype("string").fillna("__MISSING__")
            df[c] = np.where(s.isin(list(allowed)), s, rare)
        return df

    def force_category(df, cols):
        df = df.copy()
        for c in cols:
            if c in df.columns:
                df[c] = df[c].astype("category")
        return df

    def cast_bool_int8(df):
        df = df.copy()
        for c in df.select_dtypes(include=["bool"]).columns:
            df[c] = df[c].astype("int8")
        return df

    topk_maps    = fit_topk_maps(X_train_raw, HIGH_CARD_COLS)
    X_train_raw2 = apply_topk_maps(X_train_raw, topk_maps)
    X_val_raw2   = apply_topk_maps(X_val_raw,   topk_maps)
    X_test_raw2  = apply_topk_maps(X_test_raw,  topk_maps)

    CAT_COLS_FINAL = [c for c in (NATIVE_CAT_COLS + HIGH_CARD_COLS)
                      if c in X_train_raw2.columns]

    X_train_raw2 = cast_bool_int8(force_category(X_train_raw2, CAT_COLS_FINAL))
    X_val_raw2   = cast_bool_int8(force_category(X_val_raw2,   CAT_COLS_FINAL))
    X_test_raw2  = cast_bool_int8(force_category(X_test_raw2,  CAT_COLS_FINAL))

    all_cols = X_train_raw2.columns.tolist()
    num_cols = [c for c in all_cols if c not in CAT_COLS_FINAL]
    cat_cols = [c for c in CAT_COLS_FINAL if c in all_cols]

    rest = [c for c in X_train_raw2.columns if c not in num_cols and c not in cat_cols]
    if rest:
        print(f"[WARN] dropping non-typed cols: {rest}")
        X_train_raw2.drop(columns=rest, inplace=True)
        X_val_raw2.drop(columns=rest,   inplace=True)
        X_test_raw2.drop(columns=rest,  inplace=True)

    transformer = ColumnTransformer(
        transformers=[("num", Pipeline([("imp", SimpleImputer(strategy="median"))]), num_cols)],
        remainder="passthrough", verbose_feature_names_out=False
    )

    X_train = transformer.fit_transform(X_train_raw2)
    X_val   = transformer.transform(X_val_raw2)
    X_test  = transformer.transform(X_test_raw2)

    for c in CAT_COLS_FINAL:
        for X in (X_train, X_val, X_test):
            if c in X.columns:
                X[c] = X[c].astype("category")

    feature_names = X_train.columns.tolist()

    # ── XGBoost 모델 학습 ─────────────────────────────────────────────────
    # sweep config 변수명: ML_Advanced_Final과 통일
    xgb_model = XGBClassifier(
        n_estimators=config.n_estimators,
        max_depth=config.max_depth,
        learning_rate=config.learning_rate,
        subsample=config.subsample,
        colsample_bytree=config.colsample_bytree,
        reg_lambda=config.reg_lambda,
        reg_alpha=config.reg_alpha,
        min_child_weight=config.min_child_weight,
        gamma=config.gamma,
        scale_pos_weight=config.scale_pos_weight,
        tree_method="hist",
        enable_categorical=True,
        objective="binary:logistic",
        eval_metric="aucpr",
        random_state=SEED,
        n_jobs=-1,
    )
    xgb_model.fit(X_train, y_train)

    # ── Edge-level 평가 ───────────────────────────────────────────────────
    prob_val  = xgb_model.predict_proba(X_val)[:, 1]
    prob_test = xgb_model.predict_proba(X_test)[:, 1]

    val_roc,  val_auprc  = safe_auc(y_val,  prob_val)
    test_roc, test_auprc = safe_auc(y_test, prob_test)

    wandb.log({
        "val_roc_auc": val_roc, "val_auprc": val_auprc,
        "test_roc_auc": test_roc, "test_auprc": test_auprc,
        "n_train": len(y_train), "pos_train": int(y_train.sum()),
        "pos_val": int(y_val.sum()), "pos_test": int(y_test.sum()),
    })

    # PR Curve 로깅
    for split_name, y_true, prob in [("val", y_val, prob_val), ("test", y_test, prob_test)]:
        prec, rec, _ = precision_recall_curve(y_true, prob)
        _, ap = safe_auc(y_true, prob)
        fig = plt.figure(figsize=(5, 4))
        plt.plot(rec, prec, lw=1.5)
        plt.xlabel("Recall"); plt.ylabel("Precision")
        plt.title(f"{split_name.upper()} PR Curve (AUPRC={ap:.4f})")
        plt.tight_layout()
        wandb.log({f"{split_name}_pr_curve": wandb.Image(fig)})
        plt.close(fig)

    # ── Top-k workload (누적 기준) ─────────────────────────────────────────
    for k_pos in [450, 750, 1500, 3000]:
        N_v, p_v, r_v, f_v, found_v = workload_to_find_k_positives(y_val.values,  prob_val,  k_pos)
        N_t, p_t, r_t, f_t, found_t = workload_to_find_k_positives(y_test.values, prob_test, k_pos)
        wandb.log({
            f"val_find_{k_pos}_N_required": N_v, f"val_find_{k_pos}_precision": p_v,
            f"val_find_{k_pos}_recall": r_v,     f"val_find_{k_pos}_f1": f_v,
            f"val_find_{k_pos}_found_tp": found_v,
            f"test_find_{k_pos}_N_required": N_t, f"test_find_{k_pos}_precision": p_t,
            f"test_find_{k_pos}_recall": r_t,     f"test_find_{k_pos}_f1": f_t,
            f"test_find_{k_pos}_found_tp": found_t,
        })

    # ── Per-day KPI (하루 기준) ────────────────────────────────────────────
    log_per_day_kpi_to_wandb("val",  val_df,  y_val.values,  prob_val,  k_list=(30,50,100,200))
    log_per_day_kpi_to_wandb("test", test_df, y_test.values, prob_test, k_list=(30,50,100,200))

    # ── Distinct account Top-N ─────────────────────────────────────────────
    days_test = test_df["ts_day"].nunique()
    topN = 30 * days_test
    res = topN_distinct_account_metrics(
        df_meta=test_df, y_true=y_test.values, y_score=prob_test,
        topN=topN, account_cols=("From Account","To Account"), score_agg="max"
    )
    wandb.log({
        "test_cum_topN": res["topN_trx"],
        "test_cum_distinct_accounts": res["distinct_accounts"],
        "test_cum_tp_accounts": res["tp_accounts"],
        "test_cum_fp_accounts": res["fp_accounts"],
        "test_cum_precision_accounts": res["precision_accounts"],
    })

    # ── Score bin distribution ─────────────────────────────────────────────
    wandb.log({
        "val_score_bin":  wandb.Table(dataframe=score_bin_distribution(y_val.values,  prob_val)),
        "test_score_bin": wandb.Table(dataframe=score_bin_distribution(y_test.values, prob_test)),
    })

    # ── Account-level AUC ─────────────────────────────────────────────────
    val_acc  = aggregate_to_account_level(val_df,  y_val.values,  prob_val,
                                          ("From Account","To Account"))
    test_acc = aggregate_to_account_level(test_df, y_test.values, prob_test,
                                          ("From Account","To Account"))
    val_acc_roc,  val_acc_auprc  = safe_auc(val_acc["label"].values,  val_acc["score"].values)
    test_acc_roc, test_acc_auprc = safe_auc(test_acc["label"].values, test_acc["score"].values)
    wandb.log({
        "val_acc_roc_auc":  val_acc_roc,  "val_acc_auprc":  val_acc_auprc,
        "test_acc_roc_auc": test_acc_roc, "test_acc_auprc": test_acc_auprc,
    })

    # ── Account-level Find-K workload ─────────────────────────────────────
    for k_pos in [450, 750, 1500, 3000]:
        N_v, p_v, r_v, f_v, found_v = workload_to_find_k_positives(
            val_acc["label"].values, val_acc["score"].values, k_pos)
        N_t, p_t, r_t, f_t, found_t = workload_to_find_k_positives(
            test_acc["label"].values, test_acc["score"].values, k_pos)
        wandb.log({
            f"val_acc_find_{k_pos}_N_required": N_v, f"val_acc_find_{k_pos}_precision": p_v,
            f"val_acc_find_{k_pos}_recall": r_v,     f"val_acc_find_{k_pos}_f1": f_v,
            f"test_acc_find_{k_pos}_N_required": N_t, f"test_acc_find_{k_pos}_precision": p_t,
            f"test_acc_find_{k_pos}_recall": r_t,     f"test_acc_find_{k_pos}_f1": f_t,
        })

    # ── Per-day account KPI ────────────────────────────────────────────────
    log_per_day_account_kpi_to_wandb("val",  val_df,  y_val.values,  prob_val)
    log_per_day_account_kpi_to_wandb("test", test_df, y_test.values, prob_test)

    # ── Threshold @ recall 0.90 ───────────────────────────────────────────
    chosen_thr, _, _, _ = threshold_at_recall(y_val, prob_val, 0.90)
    y_val_pred  = (prob_val  >= chosen_thr).astype(int)
    y_test_pred = (prob_test >= chosen_thr).astype(int)

    wandb.log({
        "chosen_threshold": float(chosen_thr),
        "val_precision_thr":  float(precision_score(y_val,  y_val_pred,  zero_division=0)),
        "val_recall_thr":     float(recall_score(y_val,     y_val_pred,  zero_division=0)),
        "val_f1_thr":         float(f1_score(y_val,         y_val_pred,  zero_division=0)),
        "test_precision_thr": float(precision_score(y_test, y_test_pred, zero_division=0)),
        "test_recall_thr":    float(recall_score(y_test,    y_test_pred, zero_division=0)),
        "test_f1_thr":        float(f1_score(y_test,        y_test_pred, zero_division=0)),
    })

    # ── Confusion Matrix ──────────────────────────────────────────────────
    for split_name, y_true, y_pred, thr in [
        ("val",  y_val,  y_val_pred,  chosen_thr),
        ("test", y_test, y_test_pred, chosen_thr),
    ]:
        cm = confusion_matrix(y_true, y_pred)
        fig, ax = plt.subplots(figsize=(5, 4))
        sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", ax=ax)
        ax.set_xlabel("Predicted"); ax.set_ylabel("True")
        ax.set_title(f"{split_name.upper()} Confusion Matrix @thr={thr:.3f}")
        plt.tight_layout()
        wandb.log({f"{split_name}_confusion_matrix": wandb.Image(fig)})
        plt.close(fig)

    # ── AML 패턴 분석 ─────────────────────────────────────────────────────
    if "Pattern_Type" in val_df.columns:
        v_50_sum,  v_50_det  = get_pattern_analysis_tables(val_df,  prob_val,  0.5)
        v_ch_sum,  v_ch_det  = get_pattern_analysis_tables(val_df,  prob_val,  chosen_thr)
        t_50_sum,  t_50_det  = get_pattern_analysis_tables(test_df, prob_test, 0.5)
        t_ch_sum,  t_ch_det  = get_pattern_analysis_tables(test_df, prob_test, chosen_thr)
        t_sum_3k, t_det_3k   = get_pattern_analysis_tables_find_k(test_df, prob_test, 3000)
        wandb.log({
            "Analysis_Val/Recall_at_0.5":              v_50_sum,
            "Analysis_Val/Recall_at_Chosen":           v_ch_sum,
            "Analysis_Test/Recall_at_0.5":             t_50_sum,
            "Analysis_Test/Recall_at_Chosen":          t_ch_sum,
            "Analysis_Test/Pattern_Summary_at_Find3000": t_sum_3k,
            "Analysis_Detail/Test_Detail_at_Chosen":   t_ch_det,
            "Analysis_Detail/Pattern_Detail_at_Find3000": t_det_3k,
        })

    # ── Find-K 에러 분석 ──────────────────────────────────────────────────
    # Feature Importance
    booster   = xgb_model.get_booster()
    gain_dict = booster.get_score(importance_type="gain")
    imp_rows  = []
    for k, v in gain_dict.items():
        if isinstance(k, str) and k.startswith("f") and k[1:].isdigit():
            idx  = int(k[1:])
            name = feature_names[idx] if idx < len(feature_names) else k
        else:
            name = str(k)
        imp_rows.append((name, float(v)))
    imp_rows.sort(key=lambda x: x[1], reverse=True)

    tbl = wandb.Table(columns=["feature", "gain"])
    for name, gain in imp_rows:
        tbl.add_data(name, gain)
    wandb.log({"feature_importance_gain_table_all": tbl})

    top50 = imp_rows[:50]
    fig_imp, ax_imp = plt.subplots(figsize=(8, 12))
    names = [x[0] for x in top50][::-1]
    vals  = [x[1] for x in top50][::-1]
    colors_fi = ["gold" if n.startswith("sage_") else "steelblue" for n in names]
    ax_imp.barh(names, vals, color=colors_fi)
    ax_imp.set_xlabel("Gain")
    ax_imp.set_title("Top-50 Feature Importance (Gain)\n(gold=SAGE features)")
    plt.tight_layout()
    wandb.log({"feature_importance_gain_bar_top": wandb.Image(fig_imp)})
    plt.close(fig_imp)

    top_features = [name for name, _ in imp_rows[:10]]

    # =========================
    # (C-5.5) Top gain feature 기준 "test-case 영향" 테이블
    #   - 해석용이므로 X_test_raw2(가공 직후 raw)를 사용
    #   - prob_test / y_test 사용
    # =========================
    def build_feature_impact_tables(
        X_raw: pd.DataFrame,
        y_true: pd.Series,
        y_score: np.ndarray,
        top_features,
        threshold: float = 0.5,
        max_cat_levels: int = 15,
        high_value_quantile: float = 0.9,
    ):
        """
        반환:
          - binary_num_tbl: numeric/bool feature 영향 요약 (상위구간/flag)
          - cat_tbl: categorical feature 영향 요약 (카테고리별 top)
        """
        y_true = pd.Series(y_true).reset_index(drop=True)
        y_score = pd.Series(y_score).reset_index(drop=True)
        X_raw = X_raw.reset_index(drop=True)

        y_pred = (y_score >= threshold).astype(int)

        rows_num = []
        rows_cat = []

        for f in top_features:
            if f not in X_raw.columns:
                continue

            s = X_raw[f]

            # ---------- Categorical ----------
            if pd.api.types.is_categorical_dtype(s) or pd.api.types.is_object_dtype(s) or pd.api.types.is_string_dtype(s):
                s2 = s.astype("string").fillna("__MISSING__")

                tmp = pd.DataFrame({"x": s2, "y": y_true, "pred": y_pred})
                g = tmp.groupby("x", dropna=False).agg(
                    n=("y", "size"),
                    pos=("y", "sum"),
                    pred_pos=("pred", "sum"),
                    tp=("y", lambda v: int(((tmp.loc[v.index, "y"] == 1) & (tmp.loc[v.index, "pred"] == 1)).sum())),
                )
                # recall/precision 계산
                g["recall"] = np.where(g["pos"] > 0, g["tp"] / g["pos"], np.nan)
                g["precision"] = np.where(g["pred_pos"] > 0, g["tp"] / g["pred_pos"], np.nan)

                # 너무 희귀한 레벨은 제외(표가 너무 길어짐 방지)
                g = g.sort_values(["pos", "n"], ascending=False).head(max_cat_levels).reset_index()

                for _, r in g.iterrows():
                    rows_cat.append({
                        "feature": f,
                        "level": r["x"],
                        "n_test": int(r["n"]),
                        "pos_test": int(r["pos"]),
                        "pred_pos": int(r["pred_pos"]),
                        "tp": int(r["tp"]),
                        "recall": float(r["recall"]) if pd.notna(r["recall"]) else np.nan,
                        "precision": float(r["precision"]) if pd.notna(r["precision"]) else np.nan,
                    })
                continue

            # ---------- Numeric / Bool ----------
            # bool은 1인 케이스를 바로 보자
            if pd.api.types.is_bool_dtype(s):
                mask = s.fillna(False).astype(bool)
                cond_name = "== True"
            else:
                # 숫자형이면 상위 q 구간(기본 90%)을 “high”로 정의
                # (멘토님이 말한 "피처 케이스가 몇 건"에 가장 자연스러운 정의)
                s_num = pd.to_numeric(s, errors="coerce")
                if s_num.notna().sum() == 0:
                    continue
                qv = s_num.quantile(high_value_quantile)
                mask = s_num >= qv
                cond_name = f">= q{int(high_value_quantile*100)}({qv:.4g})"

            idx = np.where(mask.values)[0]
            if len(idx) == 0:
                continue

            y_sub = y_true.iloc[idx]
            pred_sub = y_pred.iloc[idx]

            n = len(idx)
            pos = int(y_sub.sum())
            pred_pos = int(pred_sub.sum())
            tp = int(((y_sub == 1) & (pred_sub == 1)).sum())

            recall = (tp / pos) if pos > 0 else np.nan
            precision = (tp / pred_pos) if pred_pos > 0 else np.nan

            rows_num.append({
                "feature": f,
                "condition": cond_name,
                "n_test": n,
                "pos_test": pos,
                "pred_pos": pred_pos,
                "tp": tp,
                "recall": recall,
                "precision": precision,
            })

        binary_num_tbl = pd.DataFrame(rows_num).sort_values(
            ["pos_test", "n_test"], ascending=False
        ) if len(rows_num) else pd.DataFrame(
            columns=["feature","condition","n_test","pos_test","pred_pos","tp","recall","precision"]
        )

        cat_tbl = pd.DataFrame(rows_cat).sort_values(
            ["pos_test", "n_test"], ascending=False
        ) if len(rows_cat) else pd.DataFrame(
            columns=["feature","level","n_test","pos_test","pred_pos","tp","recall","precision"]
        )

        return binary_num_tbl, cat_tbl


    # ---- top gain feature list 만들기
    topN_case = int(getattr(config, "topn_case_impact", 20)) if hasattr(config, "topn_case_impact") else 20
    top_features = [name for name, _ in imp_rows[:topN_case]]

    # ===== (1) thr = 0.5 =====
    thr_05 = 0.5
    impact_num_05, impact_cat_05 = build_feature_impact_tables(
        X_raw=X_test_raw2,
        y_true=y_test,
        y_score=prob_test,
        top_features=top_features,
        threshold=thr_05,
        max_cat_levels=15,
        high_value_quantile=0.9,
    )

    print("[DEBUG] impact_num_tbl shape:", impact_num_05.shape)
    print("[DEBUG] impact_cat_tbl shape:", impact_cat_05.shape)
    print("[DEBUG] impact_num_tbl head:\n", impact_num_05.head(5))
    print("[DEBUG] impact_cat_tbl head:\n", impact_cat_05.head(5))

    wandb.log({
        "test_case_impact_numeric_topgain_thr05": wandb.Table(dataframe=impact_num_05),
        "test_case_impact_categorical_topgain_thr05": wandb.Table(dataframe=impact_cat_05),
        "case_impact_threshold_thr05": thr_05,
    })

    # ===== (2) thr = 운영 thr (chosen_thr) =====
    thr_ops = float(chosen_thr)
    impact_num_ops, impact_cat_ops = build_feature_impact_tables(
        X_raw=X_test_raw2,
        y_true=y_test,
        y_score=prob_test,
        top_features=top_features,
        threshold=thr_ops,
        max_cat_levels=15,
        high_value_quantile=0.9,
    )

    print("[DEBUG] impact_num_tbl shape:", impact_num_ops.shape)
    print("[DEBUG] impact_cat_tbl shape:", impact_cat_ops.shape)
    print("[DEBUG] impact_num_tbl head:\n", impact_num_ops.head(5))
    print("[DEBUG] impact_cat_tbl head:\n", impact_cat_ops.head(5))

    wandb.log({
        "test_case_impact_numeric_topgain_throps": wandb.Table(dataframe=impact_num_ops),
        "test_case_impact_categorical_topgain_throps": wandb.Table(dataframe=impact_cat_ops),
        "case_impact_threshold_throps": thr_ops,
    })


    # -------------------------
    # AML 패턴 분석 로깅(4가지 케이스)
    # -------------------------

    # Val 데이터 분석
    v_50_sum, v_50_det = get_pattern_analysis_tables(val_df, prob_val, 0.5)
    v_ch_sum, v_ch_det = get_pattern_analysis_tables(val_df, prob_val, chosen_thr)

    # Test 데이터 분석 (상세 분석 포함)
    t_50_sum, t_50_det = get_pattern_analysis_tables(test_df, prob_test, 0.5)
    t_ch_sum, t_ch_det = get_pattern_analysis_tables(test_df, prob_test, chosen_thr)

    # WandB 로그 전송
    wandb.log({
        "Analysis_Val/Recall_at_0.5": v_50_sum,
        "Analysis_Val/Recall_at_Chosen": v_ch_sum,
        "Analysis_Detail/Test_Detail_at_Chosen": v_50_det,
        "Analysis_Detail/Test_Detail_at_Chosen": v_ch_det,
        "Analysis_Test/Recall_at_0.5": t_50_sum,
        "Analysis_Test/Recall_at_Chosen": t_ch_sum,
        "Analysis_Detail/Test_Detail_at_Chosen": t_50_det,
        "Analysis_Detail/Test_Detail_at_Chosen": t_ch_det # 가장 중요한 데이터
    })

    # test 데이터에 대해 3000건 적발 목표 기준 패턴 분석
    t_sum_3000, t_det_3000 = get_pattern_analysis_tables_find_k(test_df, prob_test, 3000)

    wandb.log({
        "Analysis_Test/Pattern_Summary_at_Find3000": t_sum_3000,
        "Analysis_Detail/Pattern_Detail_at_Find3000": t_det_3000
    })

    # 최악의 패턴 리포트 (선택 사항)
    # t_ch_sum 데이터프레임이 비어 있지 않다면 가장 낮은 Recall 패턴을 Summary에 기록 가능

    # 1. 분석에 사용할 상위 중요 피처 추출 (Top 10개)
    # 이미 코드 상단에서 계산된 imp_rows 활용
    top_features_for_analysis = [name for name, _ in imp_rows[:10]]

    # 2. 정탐/오탐/미탐 피처 및 일자별 분석 실행 (Test 데이터 기준)
    log_detailed_error_analysis(
        split_name="test",
        df=test_df,
        y_true=y_test.values,
        y_score=prob_test,
        threshold=chosen_thr,
        top_features=top_features_for_analysis
    )

    log_detailed_error_analysis(
        split_name="test",
        df=test_df,
        y_true=y_test.values,
        y_score=prob_test,
        threshold=0.5,
        top_features=top_features_for_analysis
    )

    # 3. 패턴 중심 에러 분석 실행
    log_pattern_error_analysis(
        split_name="test",
        df=test_df,
        y_true=y_test.values,
        y_score=prob_test,
        threshold=chosen_thr
    )

    log_pattern_error_analysis(
        split_name="test",
        df=test_df,
        y_true=y_test.values,
        y_score=prob_test,
        threshold=0.5
    )

    # 2. find-3000 기준으로 정탐/오탐/미탐 분석 실행 (Test 데이터셋)
    # 이 로직은 내부적으로 workload_to_find_k_positives를 호출하여 N_required를 자동으로 계산합니다.
    log_find_k_error_analysis(
        split_name="test",
        df=test_df,
        y_true=y_test.values,
        y_score=prob_test,
        k_pos_target=3000,
        top_features=top_features_for_analysis
    )


    print("=" * 60)
    print(f"[VAL]  ROC={val_roc:.4f}  AUPRC={val_auprc:.4f}")
    print(f"[TEST] ROC={test_roc:.4f}  AUPRC={test_auprc:.4f}")
    print(f"Chosen thr (val@recall=0.90): {chosen_thr:.4f}")
    print(classification_report(y_test, y_test_pred, zero_division=0))
    print("=" * 60)

    # Artifacts 저장
    os.makedirs("artifacts_out", exist_ok=True)
    joblib.dump(node_scaler,  "artifacts_out/node_scaler.joblib")
    joblib.dump(transformer,  "artifacts_out/xgb_transformer.joblib")
    joblib.dump(xgb_model,    "artifacts_out/xgb_model.joblib")
    torch.save(model.state_dict(), "artifacts_out/graphsage_weights.pt")

    art = wandb.Artifact("aml_2stage_bundle", type="model",
                         description="GraphSAGE + XGB 2-stage AML model")
    for f in ["artifacts_out/node_scaler.joblib",
              "artifacts_out/xgb_transformer.joblib",
              "artifacts_out/xgb_model.joblib",
              "artifacts_out/graphsage_weights.pt"]:
        art.add_file(f)
    wandb.log_artifact(art)

    # best run 저장
    global _best_run_artifacts
    val_auprc_val = val_auprc if not np.isnan(val_auprc) else 0.0
    try:
        if val_auprc_val > _best_run_artifacts.get("val_auprc", -1):
            _best_run_artifacts = {
                "val_auprc":    val_auprc_val,
                "model":        model,
                "node_scaler":  node_scaler,
                "transformer":  transformer,
                "xgb_model":    xgb_model,
                "acc2idx":      acc2idx,
                "sage_probs":   infer_graphsage(model, build_single_graph(
                    pd.concat([train_df[NODE_REQUIRED_COLS],
                               val_df[NODE_REQUIRED_COLS],
                               test_df[NODE_REQUIRED_COLS]], ignore_index=True),
                    *build_node_features(
                        pd.concat([train_df[NODE_REQUIRED_COLS],
                                   val_df[NODE_REQUIRED_COLS],
                                   test_df[NODE_REQUIRED_COLS]], ignore_index=True),
                        scaler=node_scaler, fit_scaler=False)[:2])[0]),
                "prob_val":     prob_val,
                "prob_test":    prob_test,
                "y_val":        y_val,
                "y_test":       y_test,
                "chosen_thr":   chosen_thr,
                "imp_rows":     imp_rows,
                "feature_names": feature_names,
            }
    except Exception:
        pass

    del xgb_model, transformer
    gc.collect()
    run.finish()


_best_run_artifacts = {}
print("train_eval_sweep 함수 정의 완료.")
```
